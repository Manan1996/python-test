```python
CALCULATING VOLUME SHOCK,PRICE SHOCK AND PRICE SHOCK WITHOUT VOLUME SHOCK.
```

```python
Volume_shock = []
v_Direction = []
Volume_shock.append(0)
v_Direction.append(0)

Price_shock = []
p_Direction = []
Price_shock.append(0)
p_Direction.append(0)

PS_without_WS =[]
PS_without_WS.append(0)


for i in range(1,len(TCS_DATA)):
    if TCS_DATA['Volume'][i] > (TCS_DATA['Volume'][i-1]+0.1*TCS_DATA['Volume'][i-1]):
        Volume_shock.append(1);
        v_Direction.append(1);
    else:
        Volume_shock.append(0); 
        v_Direction.append(0);
    if TCS_DATA['Close'][i] > (TCS_DATA['Close'][i-1]+0.02*TCS_DATA['Close'][i-1]):
        Price_shock.append(1);
        p_Direction.append(1);
    else:
        Price_shock.append(0); 
        p_Direction.append(0);
TCS_DATA["Volume_Shock"] = pd.DataFrame(Volume_shock)
TCS_DATA["Volume_Direction"] = pd.DataFrame(v_Direction)

TCS_DATA["Price_Shock"] = pd.DataFrame(Price_shock)
TCS_DATA["Price_Direction"] = pd.DataFrame(p_Direction)

for i in range(1,len(TCS_DATA)):
    if (TCS_DATA["Price_Shock"][i]==1) and (TCS_DATA["Volume_Shock"][i]==0):
        PS_without_WS.append(1)
    else:
        PS_without_WS.append(0)
TCS_DATA["PS_without_WS"] = pd.DataFrame(PS_without_WS) 

TCS_DATA.head()
```

```python
from bokeh.plotting import figure, show,  output_notebook
import datetime
from bokeh.io import reset_output()
from bokeh.io import output_notebook()
TCS = pd.read_csv('TCS_DATA.csv',parse_dates=['Date'])


plot=figure(x_axis_label='Date', x_axis_type='datetime',y_axis_label='Closing Price')
plot.line(TCS['Date'],TCS['Close'],line_width=2,color='blue')

show(plot) #PLOT 1 -Color timeseries in simple blue color

TCS_1 = TCS_DATA.loc[TCS_DATA["Volume_Shock"]==1]
TCS_0 = TCS_DATA.loc[TCS_DATA["Volume_Shock"]==0]

TCS_1['Date'] = pd.to_datetime(TCS_1['Date'])
TCS_0['Date'] = pd.to_datetime(TCS_0['Date'])

TCS_1 = TCS_1.reindex(np.arange(TCS_1.index.min(), TCS_1.index.max() + 1))
TCS_0 = TCS_0.reindex(np.arange(TCS_0.index.min(), TCS_0.index.max() + 1))

plt = figure(x_axis_label='Date', x_axis_type='datetime',y_axis_label='Closing Price')

plt.circle(TCS_1['Date'],TCS_1['Close'],fill_color='Red')
plt.circle(TCS_0['Date'],TCS_0['Close'],color='Blue')

show(plt) #PLOT 2  Color timeseries between two volume shocks in a different color

```

```python
plt_1=figure(x_axis_label='Date', x_axis_type='datetime',y_axis_label='Closing Price')
TCS_PWV=TCS_DATA.loc[TCS_DATA["PS_without_WS"]==1]
TCS_PWV['Date'] = pd.to_datetime(TCS_PWV['Date'])

plt_1.line(TCS['Date'],TCS['Close'],line_width=2,color='blue')
plt_1.circle(TCS_PWV['Date'],TCS_PWV['Close'],size=10,fill_color='Black')
show(plt_1) #PLOT 2 -Mark closing Pricing shock without volume shock to identify volumeless price movement.
```
